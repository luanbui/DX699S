# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

Side by Side Subplots

Advantages:
- Clear Week Separation: Each week is distinctly separated making it easy to focus on individual week patterns
- Independent Scaling: Each subplot can have its own y-axis scale

Disadvantages:
- Difficult Comparison: Hard to compare the same day across different weeks
- No Trend Visualization: Cannot see if there's an overall trend from week 1 to week 2


Overlaid Lines

Advantages:
- Easy Comparison: Can directly compare the same day across weeks
- Trend Detection: Can identify if visitor patterns are improving, declining, or staying consistent

Disadvantages:
- Visual Confusion: Lines may overlap or cross making it harder to follow individual week patterns
- Color Dependency: Requires different colors/styles to distinguish weeks which may not work for colorblind users

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
years = ['2008', '2009', '2010', '2011', '2012', '2013', '2014']

# Price
data = {
    'Product A': [430, 470, 390, 400, 280, 260, 120],
    'Product B': [390, 420, 380, 390, 270, 250, 110], 
    'Product C': [180, 240, 200, 190, 160, 240, 0],
    'Product D': [170, 200, 180, 160, 270, 230, 220],
    'Product E': [100, 140, 120, 110, 90, 110, 220]
}

# Set up the figure
fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.12
x_pos = np.arange(len(products))

# Colors for each year (using a colorblind-friendly palette)
colors = ['#8DA0CB', '#FC8D62', '#66C2A5', '#E78AC3', '#A6D854', '#FFD92F', '#E5C494']

# Create bars
for i, year in enumerate(years):
    values = [data[product][i] for product in products]
    bars = ax.bar(x_pos + i * bar_width, values, bar_width, 
                  label=year, color=colors[i], alpha=0.9,
                  edgecolor='white', linewidth=1.5,
                  capstyle='round')

ax.set_xlabel('', fontsize=12)
ax.set_ylabel('', fontsize=12)
ax.set_title('Average Retail Product Price per Year', fontsize=14, fontweight='bold', pad=20)

# Set x-axis
ax.set_xticks(x_pos + bar_width * 3)
ax.set_xticklabels(products, fontsize=11)

# Set y-axis
ax.set_ylim(0, 500)
ax.set_yticks([0, 100, 200, 300, 400, 500])
ax.set_yticklabels(['$0', '$100', '$200', '$300', '$400', '$500'])

# Add horizontal gridlines
ax.grid(True, axis='y', alpha=0.3, linestyle='-', linewidth=0.5)
ax.set_axisbelow(True)

# Create legend at the bottom
ax.legend(loc='lower center', fontsize=10, ncol=7, bbox_to_anchor=(0.5, -0.1))

plt.tight_layout()
plt.show()
```
